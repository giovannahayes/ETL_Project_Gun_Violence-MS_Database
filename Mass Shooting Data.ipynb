```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
#Store CSV into Dataframe

df = "./Resources/Mass Shootings Dataset Ver 5 2.csv"
mass_shootings_data_df = pd.read_csv(df)
mass_shootings_data_df
```

```python
#Split Location Column 

split_df = mass_shootings_data_df["Location"].str.split(',',1, expand = True)
split_df
```

```python
# Convert Columns to Strings
split_df.columns = split_df.columns.map(str)
```

```python
# Rename Columns 

rename_df= split_df.rename(columns={"0": "City", "1": "State"})
rename_df.head()
```

```python
#Drop location column from mass_shootings_data_df 

mass_shootings_data_df.drop('Location', inplace=True, axis=1)
mass_shootings_data_df.head()
```

```python
# Drop Age Column 
mass_shootings_data_df.drop('Age', inplace=True, axis=1)
mass_shootings_data_df.head()
```

```python
#Create new dataframe that includes City and State 

new_df = pd.concat([mass_shootings_data_df, rename_df], axis=1, sort=False)
new_df
```

```python
new_df.to_csv(r'/Users/aliciasmith/Desktop/ETL_Project_Gun_Violence-MS_Database/MS.csv', index = False)
```

```python
# # Drop NA in Latitude Column 
# new_df.dropna(subset=['Latitude'])

new_df = new_df[new_df['Latitude'].notna()]
new_df
```

```python
new_df.isnull().sum()
```

```python
title_df = new_df[["Title", "Date", "Incident Area", "Fatalities", "Injured", "Race", "Gender", "Latitude", "Longitude", "City", "State"]]
title_df
```

```python
# Organize the columns so they are in a more logical order

final1_df = title_df[["Date", "Title", "Incident Area", "City", "State", "Latitude", "Longitude", "Fatalities", "Injured", "Race", "Gender"]]
final1_df



```

```python
# Reset index 
final1_df.reset_index(drop=True)
```

```python
#Check data types

final1_df.info(verbose=True)
```

```python
# Format column to 3 decimal points 

final1_df['Latitude']= final1_df['Latitude'].round(decimals=3)
final1_df['Longitude'] = final1_df['Longitude'].round(decimals=3)
```

```python
#Combine Latitue and Logitude into one column 

final1_df['Latitude, Longitude'] = "(" + final1_df['Latitude'].astype(str) + ", " + final1_df['Longitude'].astype(str) + ")"
final1_df.head()
```
